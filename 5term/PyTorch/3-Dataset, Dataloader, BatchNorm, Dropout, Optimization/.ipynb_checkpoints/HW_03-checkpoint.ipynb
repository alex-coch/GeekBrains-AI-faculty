```python
import torch
from sklearn.datasets.california_housing import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import torch.utils.data as Data
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import SGD
import matplotlib.pyplot as plt
```

```python
# Import Data
housedata = fetch_california_housing()
# Split training set and test set
X_train, X_test, y_train, y_test = train_test_split(housedata.data, housedata.target,
                                                    test_size = 0.25, random_state = 13)
```

```python
# Data Standardization
scale = StandardScaler()
X_train_s = scale.fit_transform(X_train)
X_test_s = scale.transform(X_test)
```

```python
# Transform training data to data table
# housedatadf = pd.DataFrame(data=X_train_s, columns = housedata.feature_names)
# housedatadf['target'] = y_train
# housedatadf
```

```python
# Turn the data set to a sheet
X_train_t = torch.from_numpy(X_train_s.astype(np.float32))
y_train_t = torch.from_numpy(y_train.astype(np.float32))
X_test_t = torch.from_numpy(X_test_s.astype(np.float32))
y_test_t = torch.from_numpy(X_test.astype(np.float32))
```

```python
# Process the training data as a data loader
train_data = Data.TensorDataset(X_train_t, y_train_t)
test_data = Data.TensorDataset(X_test_t, y_test_t)
train_loader = Data.DataLoader(dataset = train_data, batch_size = 64, 
                               shuffle = True, num_workers = 1)
```

```python
# Process the training data as a data loader
train_data = Data.TensorDataset(X_train_t, y_train_t)
test_data = Data.TensorDataset(X_test_t, y_test_t)
train_loader = Data.DataLoader(dataset = train_data, batch_size = 64, 
                               shuffle = True, num_workers = 1)
```

```python
#                
class MLPregression(nn.Module):
    def __init__(self):
        super(MLPregression, self).__init__()
        #        
        self.hidden1 = nn.Linear(in_features=8, out_features=64, bias=True)
        self.bn1 = nn.BatchNorm1d(num_features=64)  
        self.dropout1 = nn.Dropout(0.3)
        self.hidden2 = nn.Linear(64, 100)
        self.bn2 = nn.BatchNorm1d(num_features=100) #  
        self.dropout2 = nn.Dropout(0.3)
        self.hidden3 = nn.Linear(100, 32)
        self.bn3 = nn.BatchNorm1d(num_features=32) 
        self.dropout3 = nn.Dropout(0.3)
        self.predict = nn.Linear(32, 1)
        
    # Defines the path forward propagation path
    def forward(self, x):
        x = F.relu(self.bn1(self.hidden1(x)))  # F.relu(self.hidden1(x))
        x = self.dropout1(x)
        x = F.relu(self.bn2(self.hidden2(x))) #F.relu(self.hidden2(x))
        x = self.dropout2(x)
        x = F.relu(self.bn3(self.hidden3(x))) #F.relu(self.hidden3(x))
        x = self.dropout3(x)
        output = self.predict(x)
        # Output a one-dimensional vector
        return output[:, 0]
```

```python
# Output network structure
# from torchsummary import summary
# testnet = MLPregression()
# summary(testnet, input_size=(1, 8)) # Represents 1 sample, 8 characteristics per sample
```

```python
def plotit(optimizer):
    # Define an optimizer
    testnet = MLPregression()
    optimizer = optimizer(testnet.parameters(), lr = 0.01)
    loss_func = nn.MSELoss() #            
    train_loss_all = []

    #  , total EPOCH Wheel
    for epoch in range(20):
        train_loss = 0
        train_num = 0
        # Iterate the loader of the training data
        for step, (b_x, b_y) in enumerate(train_loader):
            output = testnet(b_x) # Mlp Output on training BATCH
            loss = loss_func(output, b_y) #      
            optimizer.zero_grad() # Each iterative gradient initializes 0
            loss.backward() #Nushby, calculate gradient
            optimizer.step() # Optimize using gradients
            train_loss += loss.item() * b_x.size(0)
            train_num += b_x.size(0)
        train_loss_all.append(train_loss / train_num)

        # #          
    plt.figure(figsize = (8, 6))
    plt.plot(train_loss_all, 'ro-', label = 'Train loss')
    plt.legend()
    plt.grid()
    plt.xlabel('epoch')
    plt.ylabel('Loss')
    plt.title(str(optimizer)[0:str(optimizer).find('(')])
    plt.show()
    print(f'loss {round(train_loss_all[-1], 4)}')
```

```python
from torch.optim import SGD, Adam, RMSprop


for item in (SGD, Adam, RMSprop):
    plotit(item)
```
